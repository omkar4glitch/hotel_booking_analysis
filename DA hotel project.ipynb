## data analysis and visualization

```python
#percentage of amount which is not canceled and canceled
canceled_perc=df["is_canceled"].value_counts(normalize=True)#normlize=true will return the relative frequencies of the unique values.
print (canceled_perc)
plt.title("reservation status count")
plt.bar(["not canceled","cancaled"],df["is_canceled"].value_counts(),edgecolor="k",width=0.7)
plt.show()
```

```python
plt.figure(figsize=(8,4))
axl=sns.countplot(x="hotel",hue="is_canceled",data=df)
plt.title("reservation status in diffrent hotel")
plt.xlabel("Hotels")
plt.ylabel("number of reservation")
plt.legend(["not_canceled","canceled"])
for i in axl.containers:
    axl.bar_label(i)
```

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
city_hotel=df[df["hotel"]=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)
```

**we got the percentage of cancelation.(city hotel has more cancelation compare to resort hotel i.e:- 41% where resort hotel has 27% cancelation.**)

```python
resort_hotel=resort_hotel.groupby("reservation_status_date")[["adr"]].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[["adr"]].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title("average daily rate in city & resort hotel",fontsize=30)
plt.plot(resort_hotel.index, resort_hotel['adr'],label="Resort Hotel")
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

**the line graph above shows that the price (adr) of city hotel is low than resort hotel. thats why there are less reservation in resort hotel that city hotel.**

```python
plt.figure(figsize=(16,8))
df['month']=df['reservation_status_date'].dt.month
axl=sns.countplot(x='month',hue='is_canceled',data =df)
plt.ylabel=("number of reservation")
plt.legend=(["not_canceled",'canceled'])
plt.show()
```

```python
# Extract the month from the 'reservation_status_date' column and assign it to a new 'month' column
df['month'] = df['reservation_status_date'].dt.month

# Create a new figure for the count plot with a size of 16x8 inches
plt.figure(figsize=(16, 8))

# Create a count plot using Seaborn's `countplot` function
# Set the x-axis as 'month', the hue as 'is_canceled', and the data as the DataFrame 'df'
# Use the 'bright' palette for coloring the bars
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')

# Get the legend handles and labels for customizing the legend
legend_labels, _ = ax1.get_legend_handles_labels()

# Set the title for the count plot with a size of 20
plt.title('Reservation Status Per Month', size=20)

# Set the x-axis label as 'month' and the y-axis label as 'number of reservations'
plt.xlabel('month')
plt.ylabel=("number of reservation")

# Display the count plot
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title("adr per month")
grouped_data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index()
adr_pm=sns.barplot(x="month",y="adr",data=grouped_data)
for i in adr_pm.containers:
    adr_pm.bar_label(i)
plt.show()
```

**as you can see that the january month has the highest cancelation and also the adr(average daily rate) is highest in january.**
**therefore the cost of accommmodation is solely responsible for higher cancellation.**

```python
canceled_data=df[df['is_canceled']==1]
top_10_country=canceled_data['country'].value_counts()[:10]#filtered the 10 country
plt.title('top 10 country with reservation canceled')
plt.pie(top_10_country,labels=top_10_country.index,autopct='%1.1f%%')
plt.show()
```

**portugal has the highest cancellation rate (i.e-70%).**

```python
# Group canceled reservations by reservation_status_date and calculate the average adr
cancelled_df_adr = canceled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Filter not canceled reservations and calculate the average adr
not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plot the average daily rate for both canceled and not canceled reservations
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate')
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='not_cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='cancelled')
```

**orange line = cancelled #
blue line = not cancelled**

```python
cancelled_df_adr=cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>"2016") & (cancelled_df_adr['reservation_status_date']<"2017-09")]
not_cancelled_df_adr=not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>"2016") & (not_cancelled_df_adr['reservation_status_date']<"2017-09")]

plt.title=('average daily rate')
plt.figure(figsize=(20,6))
plt.plot(not_cancelled_df_adr['reservation_status_date'],not_cancelled_df_adr['adr'],label='not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'],cancelled_df_adr['adr'],label='cancelled')
plt.show()#data from 2016 to 2017
```

**as you can see that orange line(cancellation) is more than blue line(not cancellation) . this is the proof that cancellation has direct effect on the price**
